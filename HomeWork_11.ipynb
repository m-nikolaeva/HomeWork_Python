```python
import sympy
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sympy.symbols('x')
```

```python
f = -12 * x ** 4 * sympy.sin(sympy.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
```

```python
ans = sympy.plot(f)
```

```python
f = -12 * x ** 4 * sympy.sin(sympy.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
plot = sympy.plot(f, xlim=[-9, -3], ylim=[-500, 500])
plot = sympy.plot(f, xlim=[-3, 1], ylim=[-500, 500])
plot = sympy.plot(f, xlim=[1, 9], ylim=[-500, 500])
```

```python
ans = sympy.solveset(f)
ans
```

Корни уравнения определены при помощи графика функции на отрезке [-10; 10]. 

Приблизительное значение корней уравнения: -7.7; -5.1; -1.3; 2.3; 4.4; 8.

Функция f > 0 при x < -7.7, -5.1 < x < -1.3, 2.3 < x < 4.4, x > 8.

Функция f < 0 при -7.7 < x < -5.1, -1.3 < x < 2.3, 4.4 < x < 8.
